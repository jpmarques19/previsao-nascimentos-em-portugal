```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
# Load the data
file_path = '../data/processed/births_PT_limpo.csv'
births_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
births_data.head()
```

```python
# Convert the 'Date' column to datetime
births_data['Date'] = pd.to_datetime(births_data['Date'])

# Set the 'Date' column as the index
births_data.set_index('Date', inplace=True)

# Perform seasonal decomposition
decomposition = seasonal_decompose(births_data['Births'], model='additive', period=12)

# Plot the decomposition
plt.rcParams.update({'figure.figsize': (10,10)})
decomposition.plot()
plt.show()
```
